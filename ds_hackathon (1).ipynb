mapping categorical values of player position with real position categories 

```python
def position_mapping(a):
    if a==7.0 or a==9.0 or a==10.0 or a==11.0:
        return 'f'
    elif a==3.0 or a==4.0 or a==2.0 or a==5.0:
        return 'd'
    elif a==6.0 or a==12.0 or a==8.0 or a==13.0 or a==14.0 or a==17.0 or a==16.0 or a==19.0 or a==18.0 or a==15.0:
        return 'm'
    elif a==20.0 or a==21.0 or a==22.0 or a==23.0:
        return 'o'
    elif a==0.0 or a==1.0:
        return 'g'
    
    
```

```python
df['position_1']=df.apply(lambda x:position_mapping(x['player_position_1']), axis=1)
df['position_2']=df.apply(lambda x:position_mapping(x['player_position_2']), axis=1)
```

```python
import math as Math
def round_scores(a):
    s=a-Math.floor(a)
    if 0.08<s<0.3:
        return Math.floor(a)
    elif 0.3<=s<0.7:
        t=Math.floor(a)+0.5
        return t
    elif 0.7<=s<=0.9:
        return Math.ceil(a)
    elif s==0:
        return a
    else:
        return a
```
